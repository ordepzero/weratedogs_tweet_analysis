```python
import pandas as pd
import json
```

```python
df = pd.read_csv('twitter-archive-enhanced.csv')
df.head()
```

### Setting TweetPy API configuration

```python
import tweepy

with open('credentials.txt') as json_file:  
    data = json.load(json_file)
    consumer_key = data['consumer_key']
    consumer_secret = data['consumer_secret']

    access_token = data['access_token']
    access_token_secret = data['access_token_secret']

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)

public_tweets = api.home_timeline()
for tweet in public_tweets:
    print(tweet.text)
```

```python
try:
    redirect_url = auth.get_authorization_url()
except tweepy.TweepError:
    print('Error! Failed to get request token.')
```

### Retriving tweets by id

```python
df['tweet_id'] = df['tweet_id'].astype(str)
```

```python
status = api.statuses_lookup(list(df['tweet_id'][0:100].values))

for cont_tweet_id in range(100,len(df['tweet_id']),100):
    status = status+api.statuses_lookup(list(df['tweet_id'][cont_tweet_id:cont_tweet_id+100].values))
```

### Writting JSON on file

```python
with open('tweet_json3.txt', 'w') as outfile: 
    for each in status:
        #json.dump(each._json, outfile)
        outfile.write(json.dumps(each._json)+'\n')
```
